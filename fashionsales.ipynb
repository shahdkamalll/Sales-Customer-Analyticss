```python
import pandas as pd

sales['sale_date'] = pd.to_datetime(sales['sale_date'])

today = pd.Timestamp.today().normalize()

rfm = sales.groupby('customer_id').agg(
    Recency=('sale_date', lambda x: (today - x.max()).days),
    Frequency=('sale_date', 'count'),
    Monetary=('total_amount', 'sum')
).reset_index()

print(rfm.head(40) )
```

```python
rfm.to_csv('/content/drive/MyDrive/fashionsales/RFM.csv', index = False)
```

```python
def rfm_label(Frequency):
    if 5 < Frequency <= 7:
        return "vip"
    elif Frequency == 4:
        return "Loyal"
    elif Frequency == 3:
        return "Potential"
    elif Frequency == 2:
        return "Needs Attention"
    else:
        return "At Risk"

rfm['RFM_Label'] = rfm['Frequency'].apply(rfm_label)
```
